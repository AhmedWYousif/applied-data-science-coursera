```python
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# loading Data
df_us = pd.read_csv(f'{os.getcwd()}\\data\\us.csv')

# prepareing Data
df_us['new cases'] = df_us['cases'] - df_us['cases'].shift(fill_value=0)
df_us['new deaths'] = df_us['deaths'] - df_us['deaths'].shift(fill_value=0)
df_us['date'] = pd.to_datetime(df_us['date']).dt.strftime('%d/%m')

max_new_cases_y = df_us['new cases'].max()
max_new_deaths_y = df_us['new deaths'].max()
max_new_cases_x = df_us[df_us['new cases'] == max_new_cases_y].index[0]
max_new_deaths_x = df_us[df_us['new deaths'] == max_new_deaths_y].index[0]

# plotting cases
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.plot(df_us.index, df_us['new cases'], c='b', label='new cases')
plt.plot(df_us.index, df_us['new deaths'], c='r', label='deaths')

#Labelling x ticks
plt.xticks(range(0, len(df_us),10), df_us.iloc[range(0, len(df_us),10)]['date'], rotation = '60')
#plt.yticks(range(0, len(df_michigan),7), pd.to_datetime(df_michigan.index[range(0, len(df_michigan),7)]), rotation = '60')

# plot points of highest
plt.plot([max_new_cases_x], [max_new_cases_y], marker='o', markersize=4, c="m", label='highest new cases')
plt.plot([max_new_deaths_x], [max_new_deaths_y], marker='o', markersize=4, c="k", label="highest deaths")

# adding labels
plt.xlabel('Day')
plt.ylabel('# cases')
plt.title("Coronavirus cases in the US on 2020")
plt.legend(frameon = False) 
plt.text(max_new_cases_x-5, max_new_cases_y+300, f'{max_new_cases_y} on {df_us.iloc[max_new_cases_x]["date"]}', c="b", fontsize=12)
plt.text(max_new_deaths_x-5, max_new_deaths_y+300, f'{max_new_deaths_y} on {df_us.iloc[max_new_deaths_x]["date"]}', c="r", fontsize=12)

# hide top and right boarder
for loc in ['top', 'right','left', 'bottom']:
    plt.gca().spines[loc].set_visible(False)



# loading states Data
df_us_states = pd.read_csv(f'{os.getcwd()}\\data\\us-states.csv')

# prepareing Data
df_us_states = df_us_states.groupby('state')['cases','deaths'].agg(cases=('cases', 'max'),  deaths=('deaths', 'max')).sort_values(by=['cases', 'deaths'], ascending=False)
top_us_states =  df_us_states[:10]


# plot barh for states
plt.subplot(1, 2, 2)
xvals = range(len(top_us_states))
plt.barh(xvals, top_us_states['deaths'], height = 0.3, color='r', label='total cases')
plt.barh([item+0.3 for item in xvals], top_us_states['cases'], height = 0.3, color='b' , label='total deaths')

#Labelling y ticks
plt.yticks(xvals,top_us_states.index, rotation = '45')

# adding labels
plt.xlabel('# cases')
plt.ylabel('States')
plt.title("Top infected covid-19 states")
plt.legend(frameon = False) 
# hide top and right boarder
for loc in ['top', 'right','left', 'bottom']:
    plt.gca().spines[loc].set_visible(False)


```
